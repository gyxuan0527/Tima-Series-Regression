```python
data=pd.DataFrame(data.iloc[:, 3:].values.reshape(-1, 18, 24).transpose(0, 2, 1).reshape(-1, 18), 
             columns=['AMB_TEMP','CH4','CO','NMHC','NO','NO2','NOx','O3','PM10','PM2.5','RAINFALL','RH','SO2','THC','WD_HR','WIND_DIREC','WIND_SPEED','WS_HR'])
data
```

```python
data['RAINFALL']=data['RAINFALL'].fillna(0)


data['AMB_TEMP'] = [None if ('#' in x or'*' in x or'x' in x or'A' in x or x==None) else float(x) for x in data['AMB_TEMP']]
data['CH4'] = [None if ('#' in x or'*' in x or'x' in x or'A' in x or x==None) else float(x) for x in data['CH4']]
data['CO'] = [None if ('#' in x or'*' in x or'x' in x or'A' in x or x==None)else float(x) for x in data['CO']]
data['NMHC'] = [None if ('#' in x or'*' in x or'x' in x or'A' in x or x==None)else float(x) for x in data['NMHC']]
data['NO'] = [None if ('#' in x or'*' in x or'x' in x or'A' in x or x==None) else float(x) for x in data['NO']]
data['NO2'] = [None if ('#' in x or'*' in x or'x' in x or'A' in x or x==None) else float(x) for x in data['NO2']]
data['NOx'] = [None if ('#' in x or'*' in x or'x' in x or'A' in x or x==None) else float(x) for x in data['NOx']]
data['O3'] = [None if('#' in x or'*' in x or'x' in x or'A' in x or x==None) else float(x) for x in data['O3']]
data['PM10'] = [None if ('#' in x or'*' in x or'x' in x or'A' in x or x==None) else float(x) for x in data['PM10']]
data['PM2.5'] = [None if ('#' in x or'*' in x or'x' in x or'A' in x or x==None) else float(x) for x in data['PM2.5']]
data['RAINFALL'] = [None if ('#' in x or'*' in x or'x' in x or'A' in x or x==None) else float(x) for x in data['RAINFALL']]
data['RH'] = [None if ('#' in x or'*' in x or'x' in x or'A' in x or x==None) else float(x) for x in data['RH']]
data['SO2'] = [None if ('#' in x or'*' in x or'x' in x or'A' in x or x==None) else float(x) for x in data['SO2']]
data['THC'] = [None if ('#' in x or'*' in x or'x' in x or'A' in x or x==None) else float(x) for x in data['THC']]
data['WD_HR'] = [None if ('#' in x or'*' in x or'x' in x or'A' in x or x==None) else float(x) for x in data['WD_HR']]
data['WIND_DIREC'] = [None if ('#' in x or'*' in x or'x' in x or'A' in x or x==None) else float(x) for x in data['WIND_DIREC']]
data['WIND_SPEED'] = [None if ('#' in x or'*' in x or'x' in x or'A' in x or x==None) else float(x) for x in data['WIND_SPEED']]
data['WS_HR'] = [None if ('#' in x or'*' in x or'x' in x or'A' in x or x==None) else float(x) for x in data['WS_HR']]

data['AMB_TEMP'] =(data['AMB_TEMP'].ffill()+data['AMB_TEMP'].bfill())/2
data['CH4'] =(data['CH4'].ffill()+data['CH4'].bfill())/2
data['CO'] =(data['CO'].ffill()+data['CO'].bfill())/2
data['NMHC'] =(data['NMHC'].ffill()+data['NMHC'].bfill())/2
data['NO'] =(data['NO'].ffill()+data['NO'].bfill())/2
data['NO2'] =(data['NO2'].ffill()+data['NO2'].bfill())/2
data['NOx'] =(data['NOx'].ffill()+data['NOx'].bfill())/2
data['O3'] =(data['O3'].ffill()+data['O3'].bfill())/2
data['PM10'] =(data['PM10'].ffill()+data['PM10'].bfill())/2
data['PM2.5'] =(data['PM2.5'].ffill()+data['PM2.5'].bfill())/2
data['RAINFALL'] =(data['RAINFALL'].ffill()+data['RAINFALL'].bfill())/2
data['RH'] =(data['RH'].ffill()+data['RH'].bfill())/2
data['SO2'] =(data['SO2'].ffill()+data['SO2'].bfill())/2
data['THC'] =(data['THC'].ffill()+data['THC'].bfill())/2
data['WD_HR'] =(data['WD_HR'].ffill()+data['WD_HR'].bfill())/2
data['WIND_DIREC'] =(data['WIND_DIREC'].ffill()+data['WIND_DIREC'].bfill())/2
data['WIND_SPEED'] =(data['WIND_SPEED'].ffill()+data['WIND_SPEED'].bfill())/2
data['WS_HR'] =(data['WS_HR'].ffill()+data['WS_HR'].bfill())/2


data.head(40)
```

```python
train=data[0:1464]
test=data[1464:]
```

```python
from sklearn.ensemble import RandomForestClassifier 
from sklearn.ensemble import RandomForestRegressor
```

```python
pm25_train=train['PM2.5'].values
pm25_test=test['PM2.5'].values
pm25_train_x = np.asarray([pm25_train[i:i+6] for i in range(1458)])
pm25_train_y = pm25_train[6:].astype(int)
pm25_test_x = np.asarray([pm25_test[i:i+6] for i in range(738)])
pm25_test_y = pm25_test[6:].astype(int)
forest = RandomForestClassifier(n_estimators = 100)
forest_fit = forest.fit(pm25_train_x, pm25_train_y)
pm25_pred_y = forest.predict(pm25_test_x)
mean_absolute_error(pm25_test_y, pm25_pred_y)
```
